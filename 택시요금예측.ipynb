```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/aiffel/data/trip.csv')

df.info()
```

```python
missing = df.isnull().sum()
missing_ratio = (missing / len(df)) * 100
print(pd.concat([missing, missing_ratio], axis=1, keys=["Missing Count", "Missing Ratio (%)"]))
```

```python
# 예: 승객 수 결측치를 중앙값으로 대체
df['passenger_count'] = df['passenger_count'].fillna(df['passenger_count'].median())

# 필수 정보 결측치는 행 제거
df = df.dropna(subset=['pickup_datetime', 'dropoff_datetime', 'pickup_longitude', 'pickup_latitude'])
```

```python
# 숫자형 변수의 분포 확인
# 결측치 판단 및 처리

numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    plt.figure(figsize=(10, 4))
    plt.title(f"{col} scatter")
    plt.scatter(df.index, df[col])
    plt.xlabel("Index")
    plt.ylabel(col)
    plt.show()
```

```python
# 이상치 판단 및 처리
# 승객 수: 0 또는 비현실적인 수치 (>6) 제거
# 거리: 음수 또는 100km 이상이면 이상치로 판단
# 요금: 0 이하이거나 비정상적으로 높은 요금 제거

df = df[(df['passenger_count'] > 0) & (df['passenger_count'] <= 6)]
df = df[df['trip_distance'] > 0]
df = df[df['fare_amount'] > 0]
df = df[df['fare_amount'] < 500] 
```
