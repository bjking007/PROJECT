```python
import random

class Character:
    def __init__(self, name, hp, power):
        self.name = name
        self.hp = hp
        self.power = power

    def is_alive(self):
        return self.hp > 0

    def attack(self, target):
        damage = random.randint(self.power - 2, self.power + 2)
        target.hp -= damage
        print(f"{self.name}이(가) {target.name}에게 {damage}의 데미지를 입혔습니다!")

class Player(Character):
    def __init__(self, name):
        super().__init__(name, hp=100, power=10)
        self.exp = 0

    def heal(self):
        heal_amount = random.randint(10, 20)
        self.hp += heal_amount
        print(f"{self.name}이(가) {heal_amount}만큼 체력을 회복했습니다!")

class Animal(Character):
    def __init__(self, name):
        hp = random.randint(20, 40)
        power = random.randint(5, 8)
        super().__init__(name, hp, power)

# 이 부분을 함수로 감싸기!
def game_loop():
    player = Player("헌터")
    animals = ["토끼", "늑대", "멧돼지"]

    print("동물 사냥 RPG 게임 시작! \n")

    while player.is_alive():
        enemy = Animal(random.choice(animals))
        print(f"\n야생의 {enemy.name} 등장! (HP: {enemy.hp})")

        while enemy.is_alive() and player.is_alive():
            action = input("1. 공격  2. 회복 > ")
            if action == "1":
                player.attack(enemy)
                if enemy.is_alive():
                    enemy.attack(player)
            elif action == "2":
                player.heal()
                enemy.attack(player)
            else:
                print("잘못된 입력입니다.")

            print(f"플레이어 HP: {player.hp} / 적 HP: {enemy.hp}")

        if not player.is_alive():
            print("게임 오버...")
            break
        else:
            exp = random.randint(10, 30)
            player.exp += exp
            print(f"{enemy.name} 처치! 경험치 +{exp} (총 경험치: {player.exp})")

        cont = input("계속 하시겠습니까? (y/n) > ")
        if cont.lower() != "y":
            break

    print("게임 종료!")

# 게임 실행
game_loop()
```
