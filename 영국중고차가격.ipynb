```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

cars = pd.read_csv('cars.csv')
brand = pd.read_csv('brand.csv')
cars.info()
```

```python
# title에서 브랜드 이름 추출
cars['brand'] = cars['title'].str.split().str[0]
```

```python
# 브랜드 이름 확인
print(cars['brand'].value_counts())
```

```python
cars['brand_name'] = cars['title'].str.split().str[0].str.lower()
brand['brand_name'] = brand['title'].str.lower()
```

```python
merged = pd.merge(cars, brand[['brand_name', 'country']], on='brand_name', how='left')
```

```python
merged.head()
```

```python
merged.isnull().sum()
```

```python
#결측치 현황 확인
merged.isnull().mean()
```

```python
merged['Engine'] = merged['Engine'].str.split('L').str[0].astype(float)
merged['Emission Class'] = merged['Emission Class'].str.split().str[1].astype(float)
# 문자열이었던 Engine, Emission Class를 float으로 변환
```

```python
#문자열형 변수 처리
merged['Service history'] = merged['Service history'].fillna('Unknown')

#결측치가 많은 행 제거
missing_count = merged.isnull().sum(axis=1)
merged_filtered = merged[missing_count < 3].copy()
```

```python
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler

sd = StandardScaler()
numeric_cols = ['Price', 'Mileage(miles)', 'Registration_Year', 'Previous Owners', 'Engine', 'Doors', 'Seats']

merged_filtered[numeric_cols] = sd.fit_transform(merged_filtered[numeric_cols])
```

```python
from sklearn.decomposition import PCA

X = merged_filtered[numeric_cols]
pca = PCA(n_components=0.7)
X_pca = pca.fit_transform(X)

# - 전체 정보의 70% 이상을 유지하는 최소한의 주성분(Principal Components) 추출
# - 결과:
#     - `num_components`: 유지된 차원 수
#     - `explained_variance_ratio_`: 유지된 정보 비율
```

```python
brand_counts_by_country = merged_filtered.groupby("country")["brand_name"].nunique().reset_index()
brand_counts_by_country.columns = ["country", "num_unique_brands"]
brand_counts_by_country = brand_counts_by_country.sort_values(by="num_unique_brands", ascending=False)

# 각 제조국별로 고유 브랜드 개수 산출
```
