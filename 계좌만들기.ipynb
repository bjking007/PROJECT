```python
import random

class Account:
    bank_name = 'SC은행'
    account_count = 0
    
    def __init__(self, owner, balance):
        self.owner = owner        # 인스턴스 변수
        self.balance = balance 
        self.account_number = self,generate_account_number()
        self.deposit_count = 0
        self.deposit_log = []
        self.withlaw_log = []
        Account.account_count += 1  # 계좌 개수 증가
    
    def generate_account_number(self):
        #"""3자리-2자리-6자리 형식으로 계좌번호 생성"""
        return f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"
    
    def deposit(self, amount):
       # """입금 기능 - 최소 1원 이상, 5회마다 1% 이자 지급"""
        if amount < 1:
            print(" 입금은 최소 1원 이상이어야 합니다.")
            return
        self.balance += amount
        self.deposit_count += 1
        self.deposit_log.append(amount)
        print(f"{format(amount, ',')}원이 입금되었습니다. 잔액: {format(self.balance, ',')}원")

        # 5회 입금마다 이자 1% 지급
        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            print(f"입금 5회 달성! 이자 {format(interest, ',')}원이 지급되었습니다. 총 잔액: {format(self.balance, ',')}원")
            
    def withdraw(self, amount):
        #"""출금 기능 - 잔액 이하만 가능"""
        if amount > self.balance:
            print(" 출금 금액이 잔액을 초과할 수 없습니다.")
            return
        self.balance -= amount
        self.withdraw_log.append(amount)
        print(f"{format(amount, ',')}원이 출금되었습니다. 잔액: {format(self.balance, ',')}원")
        
    @classmethod
    def get_account_num(cls):
         #"""총 계좌 개수 출력"""
         print(f"현재 계설된 계좌수: {cls.account_count}개")
        
    def display_info(self):
        #"""계좌 정보 출력"""
        print(f"은행이름: {Account.bank_name}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {format(self.balance, ',')}원")

    def deposit_history(self):
        #"""입금 내역 출력"""
        print(f"\n{self.owner} 님의 입금 내역:")
        if not self.deposit_log:
            print(" - 입금 내역이 없습니다.")
        else:
            for i, amount in enumerate(self.deposit_log, 1):
                print(f"  {i}. +{format(amount, ',')}원")

    def withdraw_history(self):
        #"""출금 내역 출력"""
        print(f"\n{self.owner} 님의 출금 내역:")
        if not self.withdraw_log:
            print(" - 출금 내역이 없습니다.")
        else:
            for i, amount in enumerate(self.withdraw_log, 1):
                print(f"  {i}. -{format(amount, ',')}원")
                



    

        
```
